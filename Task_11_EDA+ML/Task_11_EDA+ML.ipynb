```python
new_df
```

```python
str_col['marital-status'].unique()
```

- we can convert not married to 0 and married to 1

```python
dct = {
    'not married':0,
    'married':1
}
str_col['marital-status'] = str_col['marital-status'].map(dct)
str_col
```

```python
new_df = pd.concat([new_df,str_col['marital-status']],axis=1)
new_df.head()
```

```python
str_col.occupation.unique()
```

```python
occu_dummy = pd.get_dummies(df['occupation'],drop_first=True)
occu_dummy
```

```python
new_df = pd.concat([new_df,occu_dummy],axis=1)
new_df
```

```python
str_col.relationship.unique()
```

```python
relation_dummy = pd.get_dummies(str_col['relationship'],drop_first=True)
relation_dummy
```

```python
new_df = pd.concat([new_df,relation_dummy],axis=1)
new_df
```

```python
str_col.race.unique()
```

```python
race_dummy = pd.get_dummies(str_col['race'],drop_first=True)
race_dummy
```

```python
new_df = pd.concat([new_df,race_dummy],axis=1)
new_df.head()
```

```python
dct = {
    'Female':0,
    'Male':1
}
str_col['gender'] = str_col['gender'].map(dct)
str_col.head()
```

```python
new_df = pd.concat([new_df,str_col['gender']],axis=1)
new_df.head()
```

```python
str_col['native-country'].unique()
```

```python
country_dummy = pd.get_dummies(str_col['native-country'],drop_first=True)
country_dummy
```

```python
new_df = pd.concat([new_df,country_dummy],axis=1)
new_df
```

```python
dct = {
    '<=50K':0,
    '>50K':1
}
str_col.income = str_col.income.map(dct)
str_col.head()
```

```python
new_df = pd.concat([new_df,str_col['income']],axis=1)
new_df
```

```python
new_df.to_csv('scaled_data.csv')
```

```python
X = new_df.drop('income', 1)
y = new_df['income']
```

```python
# spliting the dataset
from sklearn.model_selection import train_test_split
Xtrain,xtest,Ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(Xtrain, Ytrain)
```

```python
y_pred = lr.predict(xtest)
y_pred
```

```python
from sklearn.metrics import*
```

```python
accuracy_score(ytest,y_pred)
```

```python
confusion_matrix(ytest,y_pred)
```

```python
sns.heatmap(confusion_matrix(ytest,y_pred),annot=True)
```

### Hyper tuning Parameter

```python
paralist = {
    'penalty':['l1','l2'],
    'tol':[1e-3,1e-4,1e-5],
    'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'multi_class' : ['auto', 'ovr', 'multinomial'],
    'max_iter':[1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000]
}
```
